```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
y = [2000,2001,2002,2003,2004,2005]
p = [1.29,1.39,1.45,1.49,1.58,1.65]
```

```python
## line plot


plt.plot(y,p)
plt.show()
```

```python
plt.plot(y,p)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population graph')
plt.show()
```

```python
## Scatter plot
plt.scatter(y,p)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population graph')
plt.show()
```

```python
plt.plot(y,p,color = 'black')
plt.scatter(y,p,color = 'red')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population graph')
plt.show()
```

```python
## Bar plot


plt.bar(y,p,color = 'pink')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population graph')
plt.show()
```

```python
plt.barh(y,p,color = 'red')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population graph')
plt.show()
```

```python
plt.plot(y,p,color = 'Black')
plt.scatter(y,p,color = 'green')
plt.bar(y,p,color = 'pink')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population graph')
plt.show()
```

```python
## histogram

plt.hist(y,color = 'pink')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population graph')
plt.show()
```

```python
import random
```

```python
points = []

for i in range(10000):
    points.append(random.normalvariate(1,3))
plt.hist(points)
plt.show()
```

```python
points = []

for i in range(10000):
    points.append(random.normalvariate(1,3))
plt.hist(points,bins=1000)
plt.show()
```
